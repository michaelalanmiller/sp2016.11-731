```python
import pattern.de
import pattern.en
import nltk.help
```

```python
# Create pairs of sentence postags
corpuspath = '/home/michael/school/11-731/sp2016.11-731/hw1/data/dev-test-train.de-en'
de_sents = []
en_sents = []

with open(corpuspath, 'r') as f:
    for row in f.read().splitlines():
        de = row.split('|||')[0]
        en = row.split('|||')[1]
        de_sents.append(de)
        en_sents.append(en)

de_pos = []
en_pos = []

for i, de in enumerate(de_sents):
    sent_with_pos = pattern.de.parse(de, tokenize=False, chunks=False)
    words = sent_with_pos.split(' ')
    pos_sent = []
    for word in words:
        pos = word.split('/')[1]
        pos_sent.append(pos)
    de_pos.append(pos_sent)
    if i % 1000 == 0:
        print(str(i/1000)+'k', pos_sent)
    #if i % 100 == 0:
        #print i, pos_sent

en_pos = []
penn_tags = ['CC', 'CD', 'DT', 'EX', 'FW', 'IN', 'JJ', 'JJR', 'JJS', 'LS', 'MD', 'NN', 'NNS', 'NNP', 'NNPS', 'PDT', 'POS' 'PRP', 'PRP$', 'RB', 'RBR', 'RBS', 'RP', 'SYM', 'TO', 'UH', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'WDT', 'WP', 'WP$', 'WRB']

for i, en in enumerate(en_sents):
    sent_with_pos = pattern.en.parse(en, tokenize=True, chunks=False)
    words = sent_with_pos.split(' ')
    pos_sent = []
    for word in words:
        pos = word.split('/')[1]
        if pos in penn_tags:
            pos_sent.append(pos)
        else: # proper name
            pos = 'NNP'
    en_pos.append(pos_sent)
    if i % 10000 == 0:
        print(str(i/1000)+'k', pos_sent)
    #if i % 100 == 0:
        #print i, pos_sent

print len(en_pos)
```

```python
# Write postags to a file

posfilepath = '/home/michael/school/11-731/sp2016.11-731/hw1/data/pos.de-en'

with open(posfilepath, 'w') as out:
    for i in range(len(de_pos)):
        line = ' '.join(de_pos[i]) + ' ||| ' + ' '.join(en_pos[i]) + '\n'
        out.write(line)
```
